# **DELTA LAKE TABLES**

```sql
CREATE TABLE bronze_lakehouse.delta_table
(
    id INT,
    name STRING,
    salary STRING,
    dept STRING
)
```

```sql
INSERT INTO bronze_lakehouse.delta_table 
VALUES 
(1,'abc','100','IT'),
(2,'xyz','100','IT'),
(3,'eee','100','IT')
```

```sql
SELECT * FROM bronze_lakehouse.delta_table
```

**DML Operations**

```sql
UPDATE bronze_lakehouse.delta_table 
SET dept = 'HR' where id = 3
```

```sql
DELETE FROM bronze_lakehouse.delta_table
```

### **DATA VERSIONING**

```sql
DESCRIBE HISTORY bronze_lakehouse.delta_table
```

### **TIME TRAVEL**

```sql
RESTORE bronze_lakehouse.delta_table TO VERSION AS OF 2
```

### **STREAMING QUERY**

```sql
%%pyspark

df = spark.readStream.table("bronze_lakehouse.delta_table")
```

```sql
%%pyspark
df.writeStream.format("delta")\
        .option("checkpointLocation","Files/streamingdata")\
        .option("path","Files/streamingdata")\
        .toTable("bronze_lakehouse.streaming_table")
```

```sql
SELECT * FROM bronze_lakehouse.streaming_table
```
